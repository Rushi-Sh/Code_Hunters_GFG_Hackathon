```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/city_hour.csv",header = 'infer',on_bad_lines='skip')

# Treating missing values of the fiels NO2
mean1=np.mean(df.loc[~df["NO2"].isna(),"NO2"].values)
df.loc[df["NO2"].isna(),"NO2"]=mean1

# Treating missing values of the fiels NOx
mean2=np.mean(df.loc[~df["NOx"].isna(),"NOx"].values)
df.loc[df["NOx"].isna(),"NOx"]=mean2

# NO2 / NOx : Good quality -: 0-40 ug/m3,Moderate quality -: 41-80 ug/m3,Poor quality -: >80 ug/m3



# Treating missing values of the fiels CO
mean3=np.mean(df.loc[~df["CO"].isna(),"CO"].values)
df.loc[df["CO"].isna(),"CO"]=mean3

#  CO : Good quality -: 0-1 mg/m3,Moderate quality -: 1-2 mg/m3,Poor quality -: >2 mg/m3


# Treating missing values of the fiels SO2
mean4=np.mean(df.loc[~df["SO2"].isna(),"SO2"].values)
df.loc[df["SO2"].isna(),"SO2"]=mean4

# SO2 : Good quality -: 0-40 ug/m3,Moderate quality -: 41-80 ug/m3,Poor quality -: >80 ug/m3


# Treating missing values of the fiels O3
mean5=np.mean(df.loc[~df["O3"].isna(),"O3"].values)
df.loc[df["O3"].isna(),"O3"]=mean5

# O3 : Good quality -: 0-50 ug/m3,Moderate quality -: 51-100 ug/m3,Poor quality -: >100 ug/m3

# Treating missing values of the fiels Benzene
mean6=np.mean(df.loc[~df["Benzene"].isna(),"Benzene"].values)
df.loc[df["Benzene"].isna(),"Benzene"]=mean6

# Benzene : Good quality -: 0-5 ug/m3,Moderate quality -: 6-10 ug/m3,Poor quality -: >10 ug/m3

# Treating missing values of the fiels Toluene
mean7=np.mean(df.loc[~df["Toluene"].isna(),"Toluene"].values)
df.loc[df["Toluene"].isna(),"Toluene"]=mean7

# Toluene : Good quality -: 0-5 ug/m3,Moderate quality -: 6-10 ug/m3,Poor quality -: >10 ug/m3

# Treating missing values of the fiels Xylene
mean8=np.mean(df.loc[~df["Xylene"].isna(),"Xylene"].values)
df.loc[df["Xylene"].isna(),"Xylene"]=mean8

# Xylene : Good quality -: 0-5 ug/m3,Moderate quality -: 6-10 ug/m3,Poor quality -: >10 ug/m3

# Treating missing values of the fiels AQI
mean9=np.mean(df.loc[~df["AQI"].isna(),"AQI"].values)
df.loc[df["AQI"].isna(),"AQI"]=mean9

# AQI : Good quality -: 0-50 ,Moderate quality -: 51-100 ,Poor quality -: 101-200,Very Poor quality -: 201-300, Severe quality -: 301-500

# 1) NO2

y1 = df.loc[:,"NO2"].values
x1 = []
count = 0
for i in range(len(df)):
  count = count+1
  x1.append(count)




plt.bar(x1,y1,color='m')
plt.xlabel("No. of Obersevations From 2015 to 2020")
plt.ylabel("Âµg/m3")
plt.title('NO2 ANALYSIS')


g1 = []
m1 = []
p1 = []

for i in range(len(df)):
  if (0<=df.loc[i,"NO2"]<=40):
    g1.append(df.loc[i,"NO2"])
  elif (41<=df.loc[i,"NO2"]<=80):
    m1.append(df.loc[i,"NO2"])
  else:
    p1.append(df.loc[i,"NO2"])


count1 = 0
c1 = []
for i in range(16006):
  count1=count1+1
  c1.append(count1)

count2 = 0
c2 = []
for i in range(24200):
  count2=count2+1
  c2.append(count2)

count3 = 0
c3 = []
for i in range(7986):
  count3=count3+1
  c3.append(count3)

```
